```python
# Student Marks Management System
# Subject: Python Programming
# Semester: 1

def calculate_grade(average):
    """Function to determine grade based on average marks."""
    if average >= 90:
        return "A+"
    elif average >= 80:
        return "A"
    elif average >= 70:
        return "B"
    elif average >= 60:
        return "C"
    elif average >= 50:
        return "D"
    else:
        return "Fail"

students = []

def add_student():
    """Function to add student details."""
    name = input("Enter student name: ")
    marks = []
    for j in range(3):
        mark = float(input(f"Enter marks for subject {j+1}: "))
        marks.append(mark)

    total = sum(marks)
    average = total / 3
    grade = calculate_grade(average)

    student = {
        "Name": name,
        "Marks": marks,
        "Total": total,
        "Average": round(average, 2),
        "Grade": grade
    }
    students.append(student)
    print(f"\n‚úÖ Student '{name}' added successfully!\n")

def display_report():
    """Function to display all student details."""
    if not students:
        print("\nNo student data available.\n")
        return

    print("\n===== Student Report =====")
    print("Name\t\tTotal\tAverage\tGrade")
    for s in students:
        print(f"{s['Name']}\t\t{s['Total']}\t{s['Average']}\t{s['Grade']}")
    print()

def save_to_file():
    """Function to save report to a text file."""
    with open("student_report.txt", "w") as file:
        for s in students:
            file.write(f"{s['Name']} - Total: {s['Total']} - Average: {s['Average']} - Grade: {s['Grade']}\n")
    print("üìÅ Report saved to 'student_report.txt'\n")

# Main program loop
while True:
    print("=== Student Marks Management System ===")
    print("1. Add Student")
    print("2. View Report")
    print("3. Save to File")
    print("4. Exit")
    
    choice = input("Enter your choice (1-4): ")

    if choice == '1':
        add_student()
    elif choice == '2':
        display_report()
    elif choice == '3':
        save_to_file()
    elif choice == '4':
        print("Exiting program... Thank you!")
        break
    else:
        print("Invalid choice! Please try again.\n")
```
